**Description**: This code calculates the sharpness of the HR images based on the maximum overall contrast. 

**Running the code**: Just add the images into the respective dirs and run the code normally.   
   


```python
import sys
import cv2
import numpy as np
import argparse
from matplotlib import pyplot as plt
import glob
```

```python
# The dirs where the images should be.
dataset = ['condoaerial', 'massachbuildings', 'ships', 'ufsm-flame', 
                   'catsfaces', 'dogsfaces',
                   'flowers', 'plantpat', 
                   'melanomaisic', 'structretina', 
                   'amazonia1', 'cbers4a', 'deepglobe', 'isaid']

def short_name(s):
    ind=s.rfind("/")
    #new_s=s[ind+1:-4]
    new_s=s[ind+1:]
    return new_s

def find_tech(s,k):
    ind = s.rfind(k)
    return ind
    
def find_contrast(value,tech):
    contrast = []
    for d in dataset:
        for f in glob.iglob('./'+d+'/'+value+'/*.*'):
            img = []
            img = cv2.imread(f, 0)
            print(short_name(f))
            incont = find_tech(short_name(f),tech) 
            if incont != -1:
                print('Found - {}!'.format(tech))
                print()
                contrast.append(img.std())
    return contrast       
```

```python
dan_high = find_contrast('High','DAN')

moesr_high = find_contrast('High','MoESR')

x = range(1,len(dan_high)+1)

plt.figure(figsize=(10,5))
plt.title("Contrast: Highest MANIQA Scores")
plt.plot(x,dan_high, color='blue', marker = 'o', label="DAN")
plt.plot(x,moesr_high,color='red', marker = 'o', label="MoESR")
plt.xticks(range(1,len(x)+1))
plt.xlabel("Image")
plt.ylabel("Contrast")
plt.legend()
plt.savefig("Contrast_High.png")
plt.show()

```

```python
dan_low = find_contrast('Low','DAN')

moesr_low = find_contrast('Low','MoESR')

x = range(1,len(dan_low)+1)

plt.figure(figsize=(10,5))
plt.title("Contrast: Lowest MANIQA Scores")
plt.plot(x,dan_low, color='blue', marker = 'o', label="DAN")
plt.plot(x,moesr_low, color='red', marker = 'o', label="MoESR")
plt.xticks(range(1,len(x)+1))
plt.xlabel("Image")
plt.ylabel("Contrast")
plt.legend()
plt.savefig("Contrast_Low.png")
plt.show()
```
